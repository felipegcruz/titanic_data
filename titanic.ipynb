# How to filter specific rows from a DataFrame

```python
above_35 = titanic[titanic["Age"]>35]
above_35.head()
```

### I’m interested in the Titanic passengers from cabin class 2 and 3

```python
class_condition = titanic["Pclass"].isin([2, 3])
survived_condition = titanic["Survived"] == 1
sex_condition = titanic["Sex"] == "male"
class_2_and_3 = titanic.loc[class_condition & survived_condition & sex_condition, :]
class_2_and_3.head(50)
```

```python
cabin_no_na = titanic[(titanic["Cabin"].notna()) & (titanic["Age"].notna())]
cabin_no_na.head(10)
```

### passangers older than 35yo

```python
adult_names = titanic.loc[titanic["Age"] > 35, "Name"]
adult_names
```
